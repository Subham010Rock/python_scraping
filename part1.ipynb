```python
from bs4 import BeautifulSoup
soup=BeautifulSoup()
import pandas as pd
import requests
import urllib.request,urllib.parse,urllib.error

product_url=[]
product_name=[]
product_price=[]
reviews=[]
ratings=[]
for page in range(1,21):
  fhand=urllib.request.urlopen("https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}".format(page))
  soup=BeautifulSoup(fhand,'html.parser')
  products=soup.find_all('div',{'data-component-type':'s-search-result'})
  for product in products:
    if product.find('h2'):
      product_name.append(product.find('h2').text)
    else:
      product_name.append("NA")
    if product.find('a',class_="a-link-normal s-no-outline"):
      product_url.append("https://www.amazon.in"+product.find('a',class_="a-link-normal s-no-outline").get('href'))
    else:
      product_url.append("NA")
    if product.find('span',class_="a-offscreen"):
      product_price.append(product.find('span',class_="a-offscreen").get_text())
    else:
      product_price.append("NA")
    if product.find('span',class_="a-size-base s-underline-text"):
      reviews.append(product.find('span',class_="a-size-base s-underline-text").text)
    else:
      reviews.append("NA")
    if product.find('span',class_="a-icon-alt"):
      ratings.append(product.find('span',class_="a-icon-alt").text)
    else:
      reviews.append("NA")
```

```python
from google.colab import files
df=pd.DataFrame({'product_name':product_name,'product_url':product_url,'product_price':product_price,'reviews':reviews,'ratings':ratings})
df.to_csv('part1.csv')
files.download('part1.csv')
```
